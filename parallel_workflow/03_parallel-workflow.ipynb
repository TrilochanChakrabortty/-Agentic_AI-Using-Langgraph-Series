```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs:  int
    balls: int
    fours: int
    sixes: int
    
    sr: float
    boundary_percentage: float
    bound_per_ball: float
    summary: str
```

```python
def calculate_sr(state: BatsmanState) -> BatsmanState:
    sr = [state['runs']/state['balls']]*100
    return {'sr': sr}
```

```python
def calculate_bound_per_ball(state: BatsmanState) -> BatsmanState:
    bpb = state['runs']/[state['fours'] + state['sixes']]
    return {'bound_per_ball': bpb}
```

```python
def calculate_boundary_percentage(state: BatsmanState) -> BatsmanState:
    boundary_percent = [(state['fours'] * 4) + (state['sixes']*6) / state['runs']]*100
    return {'boundary_percentage': boundary_percent}
```

```python
def summary(state: BatsmanState) -> BatsmanState:
    summary = f"""
strike_rate = {state['sr']} \n
ball_per_bpundary = {state['bound_per_ball']} \n
boundary_percentage = {state['boundary_percentage']}
"""

    state['summary'] = summary
    return state
```

```python
graph = StateGraph(BatsmanState)

graph.add_node("calculate_sr", calculate_sr)
graph.add_node("calculate_boundary_percentage", calculate_boundary_percentage)
graph.add_node("calculate_bound_per_ball", calculate_bound_per_ball)
graph.add_node("summary", summary)

graph.add_edge(START, "calculate_sr")
graph.add_edge(START, "calculate_boundary_percentage")
graph.add_edge(START, "calculate_bound_per_ball")

graph.add_edge("calculate_sr", "summary")
graph.add_edge("calculate_boundary_percentage", "summary")
graph.add_edge("calculate_bound_per_ball", "summary")

graph.add_edge("summary", END)

workflow = graph.compile()
```

```python
initial_state = {
    'runs' : 100,
    'balls' : 50,
    'fours' : 8,
    'sixes' : 6
}
workflow.invoke(initial_state)
```
